```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
#%run -i utils.py
import seaborn as sns
#import datetime 

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)


import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score, mean_squared_error
```

```python
df_2017 = pd.read_csv('../data/properties_2017.csv')
```

```python
prop_types = pd.read_excel('../data/zillow_data_dictionary.xlsx', 
                           sheet_name='PropertyLandUseTypeID', 
                           names=['propertylandusetypeid', 'prop_description'])

df_2017 = df_2017.join(
    prop_types.set_index('propertylandusetypeid'),
    on='propertylandusetypeid')
```

```python
df = df_2017[df_2017.prop_description == 'Single Family Residential']
ct_2017 = len(df)
completeness = pd.DataFrame(df.columns, columns=['col_name'])
completeness['type'] = completeness.apply(lambda r: df[r.col_name].dtypes, axis=1)
completeness['%_null'] = completeness.apply(lambda r: "{:.2f}".format(df[r.col_name].isnull().sum() / ct_2017), axis=1)
```
